### Imports necessários para execução

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from numpy import pi
```

### Inicializando variáveis (qubit e registro)

```python
qreg_q = QuantumRegister(1, 'q')
creg_c = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

### Aplicando as portas no circuito de forma incremental e visualizando o resultado

```python
circuit.h(qreg_q[0])
circuit.draw("mpl")
```

### Extrai vertor de estado

```python
psi = Statevector.from_instruction(circuit)
print("Vetor de estado:", psi)
```

### Visualização na esfera de Bloch

```python
display(plot_bloch_multivector(circuit))
```

```python
circuit.s(qreg_q[0])
circuit.draw("mpl")
```

```python
psi = Statevector.from_instruction(circuit)
print("Vetor de estado:", psi)
```

```python
display(plot_bloch_multivector(circuit))
```

```python
circuit.y(qreg_q[0])
circuit.draw("mpl")
```

```python
psi = Statevector.from_instruction(circuit)
print("Vetor de estado:", psi)
```

```python
display(plot_bloch_multivector(circuit))
```

```python
circuit.rx(pi / 4, qreg_q[0])
circuit.draw("mpl")
```

```python
psi = Statevector.from_instruction(circuit)
print("Vetor de estado:", psi)
```

```python
display(plot_bloch_multivector(circuit))
```

```python
circuit.ry(pi, qreg_q[0])
circuit.draw("mpl")
```

```python
psi = Statevector.from_instruction(circuit)
print("Vetor de estado:", psi)
```
