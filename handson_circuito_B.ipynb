### Imports necessários para execução

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
from qiskit.quantum_info import Statevector
from numpy import pi
from qiskit.visualization import plot_distribution, plot_histogram
import matplotlib.pyplot as plt
from qiskit_aer import AerSimulator
```

### Inicializando variáveis (qubit e registro)

```python
qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

### Aplicando as portas no circuito de forma incremental e visualizando o resultado

```python
circuit.x(qreg_q[0])
circuit.h(qreg_q[0])
circuit.cx(qreg_q[0], qreg_q[1])
circuit.h(qreg_q[0])
circuit.ry(pi, qreg_q[1])
circuit.draw("mpl")
```

### Extrai vetor de estado e apresenta com diferentes formas de viasualização

```python
# Obterm o statevector
statevector = Statevector(circuit)

probabilities = statevector.probabilities_dict()

# Get the figure object and display it
fig = plot_distribution(probabilities, figsize=(10, 6))
plt.show()

print("Theoretical probabilities:")
for state, prob in probabilities.items():
    print(f"|{state}⟩: {prob:.4f} ({prob*100:.2f}%)")
```
